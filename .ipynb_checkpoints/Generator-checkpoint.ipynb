```python
import pdfplumber

pdf_path = [
    "/content/drive/MyDrive/HP1 - Harry Potter and the Sorcerer_s Stone (1).pdf"  ,
    "/content/drive/MyDrive/HP2 - Harry Potter and the Chamber of Secrets.pdf",
    "/content/drive/MyDrive/HP3 - Harry Potter and the Prisoner of Azkaban.pdf",
    "/content/drive/MyDrive/HP4 - Harry Potter and the Goblet of Fire.pdf",
    "/content/drive/MyDrive/HP5 - Harry Potter and the Order of the Phoenix.pdf",
    "/content/drive/MyDrive/HP6 - Harry Potter and the Half-Blood Prince.pdf",
    "/content/drive/MyDrive/HP7 - Harry Potter and the Deathly Hallows.pdf"
]
textl= []

for i in range(len(pdf_path)):
  text=""
  with pdfplumber.open(pdf_path[i]) as pdf:
      for page in pdf.pages:
          page_text = page.extract_text()
          if page_text:  # some pages might be empty
              text += page_text + "\n"
  textl.append(text)
  print(f"done with book {i+1}")

print(textl[0][:500])  # preview first 500 characters
```

```python
import re
for i in range(len(textl)) :
# assume 'text' contains your full extracted PDF text
# this regex looks for variations like "CHAPTER 1", "Chapter One", etc.
  text= textl[i]
  match = re.search(r'(CHAPTER\s*1|CHAPTER\s*I|Chapter\s*1|Chapter\s*One)', text, re.IGNORECASE)

  if match:
      start_index = match.start()       # find where chapter 1 starts
      text = text[start_index:]         # slice from there onward
      print("✅ Text trimmed to start from:", text[:200])
  else:
      print(f"⚠️ Couldn't find 'Chapter 1' — check spelling or try different regex")
  textl[i]= text
```

```python
import re

for text in textl:
  # remove patterns like "Page 12" or standalone numbers between paragraphs
  text = re.sub(r'Page\s*\d+', '', text)   # remove 'Page 23'
  text = re.sub(r'\b\d+\b', '', text)      # remove isolated numbers (page numbers, etc.)

  # remove chapter headings like "CHAPTER 2", "Chapter Two", "CHAPTER II"
  text = re.sub(r'CHAPTER\s+[A-Za-z0-9IVXLC]+', '', text, flags=re.IGNORECASE)
  text = re.sub(r'Chapter\s+[A-Za-z0-9IVXLC]+', '', text, flags=re.IGNORECASE)

  # clean up extra spaces/newlines left behind
  text = re.sub(r'\s+', ' ', text).strip()

print("✅ Cleaned text preview:\n", textl[0][:1000])
```

```python

text= " ".join(textl)

with open("abc.txt", "w") as file1:
  file1.write(text)

len(text)
```

```python
chars= sorted(list(set(text)))
vocab= len(chars)
print(chars, vocab, sep='\n')
```

```python
stoi= {ch:i for i, ch in enumerate(chars)}
itos= {i:ch for i, ch in enumerate(chars)}

print(stoi, itos, sep='\n')
```

```python
import numpy as np

data= np.array([stoi[ch] for ch in text])
print(data[:300])
```

```python
SQ_LEN = 200
STEP = 1

split_idx = int(0.9 * len(data))
train_data = data[:split_idx]
val_data = data[split_idx:]


print(len(data), len(train_data), len(val_data), len(train_data)+len(val_data), sep="\n")
```

```python
import tensorflow as tf
```

```python
BATCH_SIZE = 256
BUFFER_SIZE = 10000
```

```python
def create_dataset(data):
  ds= tf.data.Dataset.from_tensor_slices(data)
  ds=ds.batch(SQ_LEN+1, drop_remainder=True)
  ds=ds.map(lambda seq: (seq[:-1], seq[1:] ))
  ds=ds.shuffle(BUFFER_SIZE).batch(BATCH_SIZE, drop_remainder=True)
  return ds
```

```python
train_dataset= create_dataset(train_data)
val_dataset= create_dataset(val_data)
```

```python
for i,j in train_dataset.take(1):
  print(i.shape, j.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense


model = Sequential([
    Embedding(vocab, 256),
    LSTM(256, return_sequences=True, stateful=True,
         recurrent_initializer='glorot_uniform'),
    Dense(vocab)
])

model.summary()
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam

loss_fn = SparseCategoricalCrossentropy(from_logits=True)

model.compile(optimizer=Adam(), loss=loss_fn, metrics=['accuracy'])
```

```python
EPOCHS = 50
hlist=[]
for epoch in range(EPOCHS):
    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    # Train — keep shuffle=False for stateful model
    history=model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=1,
        shuffle=False,

    )
    hlist.append(history)
    # Reset LSTM states after each epoch
    model.layers[1].reset_states()

```

```python
import matplotlib.pyplot as plt

try:
    acc = [h.history['accuracy'][0] for h in hlist]
    val_acc = [h.history['val_accuracy'][0] for h in hlist]
except KeyError:
    print("Warning: 'accuracy' key not found. Please check your model.compile() metrics.")
    acc_key = [key for key in hlist[0].history.keys() if 'acc' in key and 'val' not in key][0]
    val_acc_key = [key for key in hlist[0].history.keys() if 'acc' in key and 'val' in key][0]
    print(f"Using keys: '{acc_key}' and '{val_acc_key}'")
    acc = [h.history[acc_key][0] for h in hlist]
    val_acc = [h.history[val_acc_key][0] for h in hlist]

loss = [h.history['loss'][0] for h in hlist]
val_loss = [h.history['val_loss'][0] for h in hlist]

epochs_ran = len(hlist)
epochs_range = range(1, epochs_ran + 1)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

ax1.plot(epochs_range, acc, 'bo-', label='Training Accuracy')
ax1.plot(epochs_range, val_acc, 'ro-', label='Validation Accuracy')
ax1.set_ylabel('Accuracy')
ax1.set_title('Training and Validation Accuracy')
ax1.legend()
ax1.grid(True)

ax2.plot(epochs_range, loss, 'bo-', label='Training Loss')
ax2.plot(epochs_range, val_loss, 'ro-', label='Validation Loss')
ax2.set_ylabel('Loss')
ax2.set_xlabel('Epoch')
ax2.set_title('Training and Validation Loss')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import tensorflow as tf

def generate_text(model, start_string, gen_length=400, temperature=1.0):
    input_eval = [stoi[c] for c in start_string]
    input_eval = tf.expand_dims(input_eval, 0)

    text_generated = []

    model.layers[1].reset_states()

    for _ in range(gen_length):
        predictions = model(input_eval)
        # Reshape the predictions to have a batch size of 1
        predictions = tf.reshape(predictions, (1, -1, vocab))
        predictions = tf.squeeze(predictions, 0)


        predictions = predictions / temperature
        predicted_id = tf.random.categorical(predictions, num_samples=1)[-1,0].numpy()


        text_generated.append(itos[predicted_id])

        input_eval = tf.expand_dims([predicted_id], 0)

    return start_string + ''.join(text_generated)




```

```python
string=""
start="harry"

for i in range(10):
  text= (generate_text(model, start_string=start, gen_length=20, temperature=0.8))
  string+=(" "+text)
  start= text.split()[-1]

print(string)
```
